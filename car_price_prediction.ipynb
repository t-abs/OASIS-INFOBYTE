```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/car data.csv")
```

```python
df.head()
```

```python
df.isnull().sum
```

```python
df.info()
```

```python
data=df.drop(columns=['Car_Name','Fuel_Type','Selling_type','Transmission'])
```

```python
data.head()
```

```python
x=data.drop(columns='Owner',axis=1)
```

```python
y=data['Owner']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_train)
```

```python
y_pred
```

```python
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(model,filled=True)
```

```python
print("Accuracy:",model.score(x_test,y_test)*100)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_train)
```

```python
y_pred
```

```python
print("Accuracy:",lr.score(x_test,y_test)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier()
rc.fit(x_train,y_train)
```

```python
y_pred=rc.predict(x_train)
```

```python
y_pred
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(rc.estimators_[0],filled=True)
```

```python
print("Accuracy:",rc.score(x_test,y_test)*100)
```
